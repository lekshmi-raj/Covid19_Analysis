```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_states = pd.read_csv("C://Users//Lekshmi//Documents//Covid_19_Analysis//rank_states.csv")
running_total = pd.read_csv("C://Users//Lekshmi//Documents//Covid_19_Analysis//running_totals.csv")
death_compare = pd.read_csv("C://Users//Lekshmi//Documents//Covid_19_Analysis//death_comparison.csv")
```

```python
# visualizations

fig,axes = plt.subplots(1,3,figsize=(18,6))
plt.suptitle("India Covid-19 Vaccines",fontsize=30,fontweight="bold",family="Times New Roman")

sns.barplot(data=top_states.head(10),x="Vacc_Cover_Pct",y="State/UTs",color="purple",ax=axes[0])
axes[0].set_xlabel("Coverage")
axes[0].set_ylabel("State")
axes[0].set_title("State Vaccine Coverage Ranking",fontsize=20,fontweight="bold",family="Times New Roman")

sns.lineplot(data=running_total.head(10),x="running_total",y="State/UTs",ax=axes[1],color="purple")
axes[1].set_title("Cumulative Vaccine Distribution",fontweight="bold",family="Times New Roman",fontsize=20)
axes[1].set_xlabel("Running Doses")
axes[1].set_ylabel("State")

sns.scatterplot(data=death_compare.head(10),x="Coverage_Pct",y="Death Ratio",ax=axes[2],color="purple")
axes[2].set_title(" Coverage vs Death Ratio",fontsize=20,fontweight="bold",family="Times New Roman")
axes[2].set_xlabel("Coverage")
axes[2].set_ylabel("Death Ratio")

plt.tight_layout()
plt.savefig("Window_Functions",dpi=300,bbox_inches="tight")
plt.show()
```
