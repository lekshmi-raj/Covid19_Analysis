```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
Summary = pd.read_csv("C://Users//Lekshmi//Documents//Covid_19_Analysis//Summary.csv")
Summary
```

# data enhancement

```python
Summary["Vax_Coverage_%"] = Summary["Coverage_Pct"]
Summary["Death_Rate_%"] = Summary["Death_Pct"]
Summary["Deaths_Absolute"] = Summary["Deaths"]

Summary["Vax_Deaths_Rate"] = Summary["Vax_Coverage_%"] / (Summary["Death_Pct"] + 1)
Summary["Protection_Score"] = Summary["Vax_Coverage_%"] * (100 - Summary["Death_Rate_%"])

Summary["Vax_Deaths_Rate"] = Summary["Vax_Deaths_Rate"].round(2)
Summary["Protection_Score"] = Summary["Protection_Score"].round(2)
```

```python
Summary
```

```python
Summary = Summary.drop(columns=["Coverage_Pct","Deaths","Death_Pct"])
```

```python
Summary.info()
```

# state performance rankings

```python
top_vaccinated = Summary.nlargest(10,"Vax_Coverage_%")[["State/UTs","Vax_Coverage_%"]]
top_vaccinated
```

```python
top_protected = Summary.nlargest(10,"Protection_Score")[["State/UTs","Protection_Score"]]
top_protected
```

```python
top_death_rate = Summary.nlargest(10,"Death_Rate_%")[["State/UTs","Death_Rate_%"]]
top_death_rate
```

# correlation and main scatter plot

```python
# correlation matrix

plt.figure(figsize=(10,8))
corr_data = Summary[["Vax_Coverage_%","Death_Rate_%","Protection_Score"]].corr()
sns.heatmap(corr_data,annot=True,cmap="RdYlGn",center=0,square=True,cbar_kws={"label":"Correlation Coefficient"})
plt.title("Vaccination VS Correlation Matrix",fontsize=20,fontweight="bold",family="Times New Roman")
plt.tight_layout()
plt.savefig("Correlation_Matrix",dpi=300,bbox_inches="tight")
plt.show()
```

```python
# scatter plot

plt.figure(figsize=(10,8))
scatter = plt.scatter(Summary["Vax_Coverage_%"],Summary["Death_Rate_%"],s=Summary["Deaths_Absolute"]/1000,c=Summary["Protection_Score"],cmap='RdYlGn',alpha=0.9)
plt.colorbar(scatter,label="Protection Score")
plt.title("India Covid_19 Vaccination VS Death Rate by States",fontsize=20,fontweight="bold",family="Times New Roman")
plt.xlabel("Vaccination_Coverage_%")
plt.ylabel("Death_Rate")
for i,state in enumerate(Summary["State/UTs"]):
    plt.annotate(state[:3],(Summary["Vax_Coverage_%"].iloc[i],Summary["Death_Rate_%"].iloc[i]))
plt.savefig("Vax_Death_Rate",dpi=300,bbox_inches="tight")
plt.show()
```

# dashboard

```python
fig,axes = plt.subplots(2,2,figsize=(16,12))
plt.suptitle("Advanced Covid Vaccination Analysis",fontsize=25,family="Times New Roman",fontweight="bold")

axes[0,0].barh(top_protected["State/UTs"],top_protected["Protection_Score"],color="gold")
axes[0,0].invert_yaxis()
axes[0,0].set_title("Top 10 Protection Scores",fontsize=20,fontweight="bold",family="Times New Roman")
axes[0,0].set_xlabel("Protection_Score")
axes[0,0].set_ylabel("State")

axes[0,1].scatter(Summary["Vax_Coverage_%"],Summary["Death_Rate_%"],s=Summary["Deaths_Absolute"]/1000,c="steelblue",alpha=0.8)
z = np.polyfit(Summary['Vax_Coverage_%'], Summary['Death_Rate_%'], 1)
p = np.poly1d(z)
axes[0,1].plot(Summary['Vax_Coverage_%'], p(Summary['Vax_Coverage_%']), "r--", alpha=0.8)
axes[0,1].set_title("Vaccination VS Death Trends",fontsize=20,fontweight="bold",family="Times New Roman")
axes[0,1].set_xlabel("Vaccination Rate")
axes[0,1].set_ylabel("Death_Rate")

axes[1,0].scatter(Summary["Vax_Coverage_%"],Summary["Protection_Score"],s=80,c=Summary["Death_Rate_%"],cmap="Blues_r",alpha=0.7)
axes[1,0].set_title("Vaccination VS Survival",fontsize=20,fontweight="bold",family="Times New Roman")
axes[1,0].set_xlabel("Vaccination")
axes[1,0].set_ylabel("Survive_Rate")

axes[1,1].hist(Summary["Protection_Score"],bins=10,color="silver",edgecolor="gold")
axes[1,1].axvline(Summary["Protection_Score"].mean(),color="red",linestyle="--",label="National Average")
axes[1,1].set_title("Protection Score Distribution",fontsize=20,fontweight="bold",family="Times New Roman")
axes[1,1].set_xlabel("Protection Score")
axes[1,1].legend()

plt.tight_layout()
plt.savefig("Day2_Analysis",dpi=300,bbox_inches="tight")
plt.show()
```

```python
final_analysis = Summary[["State/UTs","Vax_Coverage_%","Death_Rate_%","Protection_Score"]].copy()
final_analysis["Rank"] = final_analysis["Protection_Score"].rank(ascending=False).astype(int)
final_analysis.sort_values("Rank")

final_analysis.to_csv("final_analysis.csv",index=False)
print("saved successfully")
```
