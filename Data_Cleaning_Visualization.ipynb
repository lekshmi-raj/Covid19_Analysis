```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vaccine = pd.read_csv("C://Users//Lekshmi//Documents//Covid_19_Analysis//COVID-19 India Statewise Vaccine Data.csv")
cases = pd.read_csv("C://Users//Lekshmi//Documents//Covid_19_Analysis//Latest Covid-19 India Status.csv")
```

```python
vaccine["Coverage_Pct"] = vaccine["Dose 2"] / vaccine["Population"] * 100
cases["Death_Pct"] = cases["Deaths"] / cases["Total Cases"] * 100
cases["Survive_Pct"] = 100 - cases["Death_Pct"]
```

```python
vaccine["Coverage_Pct"] = vaccine["Coverage_Pct"].round(2)
```

```python
cases["Death_Pct"] = cases["Death_Pct"].round(2)
cases["Survive_Pct"] = cases["Survive_Pct"].round(2)
```

```python
cases["State/UTs"] = cases["State/UTs"].str.replace("Telengana","Telangana",case=False)
```

```python
vaccine["State/UTs"] = vaccine["State/UTs"].str.strip().str.upper()
cases["State/UTs"] = cases["State/UTs"].str.strip().str.upper()
```

```python
merged = pd.merge(vaccine[["State/UTs","Coverage_Pct"]],
                  cases[["State/UTs","Deaths","Death_Pct"]],
                  on="State/UTs")
```

```python
merged.info()
```

```python
merged.to_csv("Summary.csv",index=False)
print("saved successfully")
```

```python
fig = plt.figure(figsize=(18,12))
fig.suptitle("Covid_19_Analysis - Day 1",fontsize=30,fontweight="bold",family="Times New Roman")
# chart 1

plt.subplot(2,3,1)
top_vax = merged.nlargest(10,"Coverage_Pct")
plt.barh(top_vax["State/UTs"],top_vax["Coverage_Pct"],color="blue")
plt.title("Top 10 Vaccinated States",fontsize=20,fontweight="bold",family="Times New Roman")
plt.xlabel("Coverage Rate")
plt.ylabel("States")
plt.gca().invert_yaxis()


# chart 2

plt.subplot(2,3,2)
top_deaths = merged.nlargest(10,"Deaths")
plt.barh(top_deaths["State/UTs"],top_deaths["Deaths"]/1000,color="green")
plt.title("Top 10 Death Rated States",fontsize=20,fontweight="bold",family="Times New Roman")
plt.xlabel("Death Rate")
plt.ylabel("States")
plt.gca().invert_yaxis()


# chart 3

plt.subplot(2,3,3)
plt.scatter(merged["Coverage_Pct"],merged["Deaths"]/1000,color="red",s=15,alpha=0.7)
plt.title("Vaccination VS Deaths",fontsize=20,fontweight="bold",family="Times New Roman")
plt.xlabel("Death Rate")
plt.ylabel("Coverage Rate")


# chart 4

plt.subplot(2,3,4)
plt.hist(merged["Coverage_Pct"],color="yellow",edgecolor="black",bins=12)
plt.title("Vaccinated Distribution",fontsize=20,fontweight="bold",family="Times New Roman")
plt.xlabel("Coverage Rate")


# chart 5

plt.subplot(2,3,5)
plt.scatter(merged["Coverage_Pct"],merged["Death_Pct"],s=15,color="brown",alpha=0.7)
plt.title("Vaccination Rate VS Death Rate",fontsize=20,fontweight="bold",family="Times New Roman")
plt.xlabel("Coverage Rate")
plt.ylabel("Death Rate")
plt.tight_layout()


plt.savefig("Day1_Analysis",dpi=300,bbox_inches="tight")
plt.show()
```
